### IMPORT LIBRARY YANG DIGUNAKAN DALAM VISUALISASI DATA ###

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("steam.csv")
df.head()
```

## MEMVISUALISASIKAN DATA ##

### 1.) GRAFIK BATANG (BAR CHART) ###

#### NOTE : KARENA JUMLAH DATA YANG TERLALU BANYAK, UNTUK BAR CHART SAYA HANYA MENGAMBIL DATA DARI RANGE 0-100 ####

```python
plt.figure(figsize=(15, 15))
harga = df.loc[0:100, "price"]
publish = df.loc[0:100, "publisher"]

x = list(harga)
y = list(publish)
         
plt.title("Perbandingan Harga Game dengan Publisher", fontsize = 20)
plt.xlabel("Game Prices", fontsize = 20)
plt.ylabel("Game Publishers",fontsize = 20)
plt.bar(x,y, color="green")
plt.show()
```

#### INFORMASI YANG DIPEROLEH:
#### DARI BAR CHART DIATAS, KITA DAPAT MEMPEROLEH INFORMASI BAHWA DIANTARA DATA YANG DI INPUT, GAME DENGAN HARGA 0 DAN 4.5$ DIMILIKI OLEH SELURUH PUBLISHER DI DALAM DATA ####

## 2.) GRAFIK PIE ##

#### KARENA JUMLAH DATA YANG TERLALU BANYAK, DISINI SAYA MENGAMBIL RANGE DATASET AGAR BISA DIBENTUK KE DALAM PIE ####

#### UNTUK DATA PRICE, SAYA MENGGUNAKAN BEBERAPA DATA PRICE SEPERTI 0.99, 1.99, 3.99, 5.99, DAN 10.99 ####

```python
less_4 = df[df["price"].isin([0.99, 1.99, 3.99, 5.99, 10.99])]
```

```python
data = less_4["price"].value_counts().rename_axis("Price").reset_index(name="Jumlah_Data")
data.head()
```

#### SETELAH ITU BARU KITA BISA MEMBENTUK DATA KE DALAM PIE CHART ####

```python
plt.figure(figsize=(10,10))
plt.pie(
    data["Jumlah_Data"], 
    labels = data["Price"],
    autopct = '%f%%')
plt.title("Perbandingan Harga Games di Steam", fontsize = 10)
plt.legend()
plt.show()
```

#### INFORMASI YANG DIPEROLEH:
#### DARI PIE CHART DIATAS, KITA DAPAT MEMPEROLEH INFORMASI BAHWA DIANTARA 5 DATA PRICE YANG DI INPUT, PRICE 3.99 MEMILIKI JUMLAH GAME TERBANYAK DAN PRICE 0.99 MEMILIKI JUMLAH GAME YANG PALING SEDIKIT

#### KITA JUGA DAPAT MENAMPILKAN DATA LAINNYA, SEPERTI CONTOH KITA AMBIL DATA PUBLISHER ####

#### DATA PUBLISHER JUGA TERLALU BANYAK, JADI AGAR BISA MEMBENTUK PIE CHART SAYA MENGAMBIL BEBERAPA DATA PUBLISHER SAJA ####

```python
less_5 = df[df["publisher"].isin(["Rockstar Games", "Ubisoft", "2K", "Valve", "Capcom"])]
```

```python
data2 = less_5["publisher"].value_counts().rename_axis("Publisher").reset_index(name="Jumlah Game")
data2.head()
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data2["Jumlah Game"], 
    labels = data2["Publisher"],
    autopct = '%f%%')
plt.title("Perbandingan Publisher Games di Steam")
plt.legend()
plt.show()
```

#### INFORMASI YANG DIPEROLEH:
#### DARI PIE CHART DIATAS, KITA DAPAT MEMPEROLEH INFORMASI BAHWA DIANTARA 5 DATA PUBLISHER YANG TELAH DI INPUT, UBISOFT MENJADI PUBLISHER DENGAN JUMLAH GAME TERBANYAK DAN ROCKSTAR GAMES MENJADI YANG PALING SEDIKIT

## 3.) SCATTERPLOT ##

#### MEMBANDINGKAN HARGA GAME DENGAN PLATFORM YANG DIMILIKI ####
#### DISINI SAYA KEMBALI MEMAKAI RANGE PADA DATASET YAITU 0-2000 DATA ####

```python
pub = df.loc[0:2000, "platforms"]
price = df.loc[0:2000, "price"]            
```

```python
plt.figure(figsize=(20,20))
plt.scatter(x = pub, y = price, color = "orange")
plt.title("Harga dan Platform di Steam", fontsize = 15)
plt.xlabel("Platforms", fontsize = 20)
plt.ylabel("Harga Game", fontsize = 20)
plt.show()
```

#### INFORMASI YANG DIPEROLEH:
#### DARI SCATTERPLOT DIATAS, PLATFORM WINDOWS MEMILIKI GAME DENGAN HARGA TERMAHAL YAITU BERADA DI KISARAN HARGA 90$. SELAIN ITU, PLATFORM WINDOWS JUGA MENJADI PLATFORM DENGAN HARGA GAME YANG PALING BERVARIASI DIBANDING PLATFORM LAINNYA

## 4.) BOXPLOT ##

#### MEMBANDINGKAN HARGA GAME DENGAN PLATFORM YANG DIMILIKI MENGGUNAKAN BOXPLOT ####

#### UNTUK DATA PEMBENTUK BOXPLOT, MASIH MENGGUNAKAN DATA YANG SAMA DENGAN SCATTERPLOT DIATAS
#### X = DATA PLATFORM DARI RANGE 0-2000 
#### Y = DATA PRICE DARI RANGE 0-2000

```python
plt.figure(figsize=(10,10))
sns.boxplot(data = df, x = pub ,y = price)
plt.show()
```

#### INFORMASI YANG DIPEROLEH:
#### PLATFORM WINDOWS;LINUX MEMILIKI MEDIAN TERTINGGI YAITU BERADA DI KISARAN 14$ YANG HANYA BERJARAK TIPIS DIBANDINGKAN DENGAN 4 PLATFORM LAINNYA
