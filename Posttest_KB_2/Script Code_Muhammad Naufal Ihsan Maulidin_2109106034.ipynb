#### MEMASUKKAN DATA YANG DIPEROLEH KE DALAM DICTIONARY ####

```python
infodata = {
'attributes': [],
'object_attributes': 0,
'number_attributes': 0,
'null_attributes': 0,
'total_records': 0,
}
```

```python
infodata['attributes'] = list(df.columns.values)
infodata['object_attributes'] = list(df.select_dtypes(['object']).columns)
infodata['number_attributes'] = list(df.select_dtypes(['float64', 'int64']).columns)
infodata['null_attributes'] = list(df.loc[:, df.isnull().any()].columns)
infodata['total_records'] = df.shape[0]
```

#### PRINT HASIL INFODATA : ####

#### 1.) JUMLAH RECORD DATA ####

```python
print("JUMLAH RECORD PADA DATA :", infodata['total_records'])
```

#### 2.) NAMA ATTRIBUTES DATA ####

```python
print("NAMA SELURUH ATTRIBUTES DATA :", infodata['attributes'])
```

#### 3.) JUMLAH ATTRIBUTES DATA ####

```python
print("JUMLAH TOTAL ATTRIBUTES DATA :", len(infodata['attributes']))
print("JUMLAH ATTRIBUTES DENGAN TIPE OBJEK :", len(infodata['object_attributes']))
print("JUMLAH ATTRIBUTES DENGAN TIPE ANGKA :", len(infodata['number_attributes']))
print("JUMLAH ATTRIBUTES NULL :", len(infodata['null_attributes']))
```
