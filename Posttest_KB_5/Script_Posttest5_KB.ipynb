#### IMPORT LIBRARY YANG AKAN DIGUNAKAN

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

#### IMPORT DATASET YANG AKAN DIGUNAKAN (Menggunakan Dataset yang telah bersih)

```python
df = pd.read_csv("Employee.csv")
```

```python
df.head(10)
```

#### MENGECEK DATASET AGAR TIDAK ADA DATA BERNILAI NULL

```python
df.isnull().sum()
```

#### SPLIT DATA

#### KITA AMBIL DATA JoiningYear, PaymentTier, Age, ExperienceInCurrentDomain, DAN LeaveOrNot SEBAGAI FEATURE
#### DAN Education SEBAGAI TARGET

```python
X = df[["JoiningYear", "PaymentTier", "Age", "ExperienceInCurrentDomain", "LeaveOrNot"]]
Y = df["Education"]

## Variabel X sebagai feature dan Variabel Y sebagai Target
```

#### MEMANGGIL LIBRARY TTS (TRAIN TEST SPLIT)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=0)
```

### PEMANGGILAN ALGORITMA (KLASIFIKASI)

```python
from sklearn.neighbors import KNeighborsClassifier
```

#### MEMASUKKAN TETANGGA KNN SECARA DEFAULT YAITU 5

```python
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train, Y_train)
```
